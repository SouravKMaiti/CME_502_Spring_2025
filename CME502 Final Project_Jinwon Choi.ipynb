# Catalytic activity of 3 wt% TMOx/BNNT (TM=Cu,Co,Ni, and Fe)

```python
import pandas as pd
import matplotlib.pyplot as plt

df_3CoOx = pd.read_excel('3 CoOx@BNNT.xlsx')
df_3CuOx = pd.read_excel('3 CuOx@BNNT.xlsx')
df_3NiOx = pd.read_excel('3 NiOx@BNNT.xlsx')
df_3FeOx = pd.read_excel('3 FeOx@BNNT.xlsx')
df_BNNT = pd.read_excel('BNNT.xlsx')

plt.figure(figsize=(7,6))

plt.plot(df_3CoOx['Temperature'], df_3CoOx['CO_conversion'], 'o-', color='red', linewidth=3, markersize=12, label = '3 wt% CoOx/BNNT')
plt.plot(df_3CuOx['Temperature'], df_3CuOx['CO_conversion'], '^-', color='blue', linewidth=3, markersize=12, label = '3 wt% CuOx/BNNT')
plt.plot(df_3NiOx['Temperature'], df_3NiOx['CO_conversion'], 's-', color='green', linewidth=3, markersize=12, label = '3 wt% NiOx/BNNT')
plt.plot(df_3FeOx['Temperature'], df_3FeOx['CO_conversion'], 'p-', color='black', linewidth=3, markersize=12, label = '3 wt% FeOx/BNNT')
plt.plot(df_BNNT['Temperature'], df_BNNT['CO_conversion'], 'h-', color='orange', linewidth=3, markersize=12, label = 'BNNT')

plt.xlabel("Temperature (°C)", fontsize=24, fontweight='bold')
plt.ylabel("CO conversion (%)", fontsize=24, fontweight='bold')
plt.legend(fontsize = 30, frameon=False, prop={'weight':'bold'})
plt.xlim(30,430)
plt.xticks(fontsize=18)
plt.ylim(0,100)
plt.yticks(fontsize=18)
plt.tight_layout()

plt.savefig('Catalytic Activity of 3 TMOx@BNNNT.png', dpi=300)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_3CoOx = pd.read_csv('TPR 3 CoOx@BNNT.CSV')
df_3CuOx = pd.read_csv('TPR 3 CuOx@BNNT.CSV')
df_3NiOx = pd.read_csv('TPR 3 NiOx@BNNT.CSV')
df_3FeOx = pd.read_csv('TPR 3 FeOx@BNNT.CSV')
df_BNNT = pd.read_csv('TPR BNNT.CSV')

fig, axs = plt.subplots(5,1, figsize=(7,10), sharex=True)

axs[0].plot(df_3CoOx['Temperature'], df_3CoOx['TCD'], color='red', linewidth=3)
axs[0].text(0.01, 0.8, '3 wt% CoOx/BNNT', transform=axs[0].transAxes, fontsize = 16, fontweight='bold')
axs[0].set_ylim(-100,500)

axs[1].plot(df_3CuOx['Temperature'], df_3CuOx['TCD'], color='blue', linewidth=3)
axs[1].text(0.01, 0.8, '3 wt% CuOx/BNNT', transform=axs[1].transAxes, fontsize = 16, fontweight='bold')
axs[1].set_ylim(-100,1000)

axs[2].plot(df_3NiOx['Temperature'], df_3NiOx['TCD'], color='green', linewidth=3)
axs[2].text(0.01, 0.8, '3 wt% NiOx/BNNT', transform=axs[2].transAxes, fontsize = 16, fontweight='bold')
axs[2].set_ylim(-100,700)

axs[3].plot(df_3FeOx['Temperature'], df_3FeOx['TCD'], color='black', linewidth=3)
axs[3].text(0.01, 0.8, '3 wt% FeOx/BNNT', transform=axs[3].transAxes, fontsize = 16, fontweight='bold')
axs[3].set_ylim(-100,400)

axs[4].plot(df_BNNT['Temperature'], df_BNNT['TCD'], color='orange', linewidth=3)
axs[4].text(0.01, 0.8, 'BNNT', transform=axs[4].transAxes, fontsize = 16, fontweight='bold')
axs[4].set_ylim(-100,400)

plt.xlabel("Temperature (°C)", fontsize=24, fontweight="bold")
plt.xlim(100,800)
plt.xticks(np.arange(100,801,100), fontsize=18, fontweight="bold")
fig.text(0.01, 0.5, 'TCD signal(a.u.)', va='center', rotation='vertical', fontsize = 24, fontweight = 'bold')

plt.savefig('H2-TPR Results of 3 TMOx@BNNNT.png', dpi=300)
plt.show
```
